### base model - with all variables variables

```python
df_train=pd.read_csv("C:\\Sridhar\\AI_ML\\Algorithms\\datasets\\HomePrices\\train.csv")
df_holdout=pd.read_csv("C:\\Sridhar\\AI_ML\\Algorithms\\datasets\\HomePrices\\test.csv")
```

```python
df_train.shape,df_holdout.shape
```

```python
from sklearn.preprocessing import LabelEncoder
for i in df_train.describe(include='object').columns:
    le=LabelEncoder()
    le.fit(df_train[i].astype(str))
    t=le.transform(df_train[i].astype(str))
    df_train[i]=t
```

```python
df_train.head()
```

```python
df_train.drop('Id',axis=1,inplace=True)
```

```python
df_train.fillna(20,inplace=True)
```

```python
y=df_train['SalePrice'].values
x=df_train.drop('SalePrice',axis=1)
```

```python
x=x.values
```

```python
df_train.shape
```

```python
from numpy.linalg import inv
```

```python
beta_coeff=np.dot((np.dot(x.T,inv(np.dot(x,x.T)))),y)
```

```python
beta_coeff
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
pd.DataFrame([beta_coeff,x.columns]).T
```
