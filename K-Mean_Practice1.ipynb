```python
import numpy as np
```

```python
base_dataset=np.random.randint(0,100,(100,5))
```

```python
base_dataset
```

```python
r1=base_dataset[np.random.randint(0,100)]
```

```python
r2=base_dataset[np.random.randint(0,100)]
```

```python
r3=base_dataset[np.random.randint(0,100)]
```

```python
## find the distance from each point to random point
```

```python
d1=np.sqrt(((r1-base_dataset)*(r1-base_dataset)).sum(axis=1))
d2=np.sqrt(((r2-base_dataset)*(r2-base_dataset)).sum(axis=1))
d3=np.sqrt(((r3-base_dataset)*(r3-base_dataset)).sum(axis=1))
```

```python
distance_data=list(zip(d1,d2,d3))
```

```python
distance_data=np.array(distance_data)
```

```python
distance_data
```

```python
distance_data.argmin(axis=1)
```

```python
base_dataset[distance_data.argmin(axis=1)==0].shape,base_dataset[distance_data.argmin(axis=1)==1].shape,base_dataset[distance_data.argmin(axis=1)==2].shape
```

```python
## Find the centroid
```

```python
c1=base_dataset[distance_data.argmin(axis=1)==0]
c2=base_dataset[distance_data.argmin(axis=1)==1]
c3=base_dataset[distance_data.argmin(axis=1)==2]
```

```python
r1_c1=c1.mean(axis=0)
```

```python
r2_c2=c2.mean(axis=0)
```

```python
r3_c3=c3.mean(axis=0)
```

```python
d1=np.sqrt(((r1_c1-base_dataset)*(r1_c1-base_dataset)).sum(axis=1))
d2=np.sqrt(((r2_c2-base_dataset)*(r2_c2-base_dataset)).sum(axis=1))
d3=np.sqrt(((r3_c3-base_dataset)*(r3_c3-base_dataset)).sum(axis=1))
```

```python
distance=list(zip(d1,d2,d3))
```

```python
distance=np.array(distance)
```
