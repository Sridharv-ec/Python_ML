```python
for k in numerical_continuous:
    data=train_data[k].values
    q1=np.quantile(data,0.25)
    q3=np.quantile(data,0.75)
    iqr=q3-q1
    utv=q3+1.5*(iqr)
    ltv=q1-1.5*(iqr)
    outliers_data=[]
    for i in data :
        if i<ltv or i>utv:
            outliers_data.append(np.median(data))
        else:
            outliers_data.append(i)
    train_data[k]=outliers_data     
```
