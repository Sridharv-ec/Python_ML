```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import wordpunct_tokenize
```

```python
df=pd.read_csv("C:\\Sridhar\\AI_ML\\Algorithms\\NLP\dataset\\Amazon_Unlocked_Mobile.csv")
```

```python
df = df.sample(1000)
```

```python
df.shape
```

```python
df.head()
```

```python
x = []
for i in df['Rating']:
    if i < 3:
        x.append(0)
    else:
        x.append(1)
```

```python
df['Rating'] = x
```

```python
df.head()
```

```python
x = df['Reviews']
y = df['Rating']
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.20, random_state=0)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Fit the CountVectorizer to the training data
vect = CountVectorizer().fit(X_train)
# transform the documents in the training data to a document-term matrix
X_train_vectorized = vect.transform(X_train)
X_test_vectorized = vect.transform(X_test)
```

```python
X_train_vectorized
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_vectorized,y_train)
```

```python
from sklearn.metrics import roc_auc_score
# Predict the transformed test documents
predictions = model.predict(X_test_vectorized)

print('AUC : ',roc_auc_score(y_test,predictions) )
```

```python
predictedValuesFromModel = model.predict(X_test_vectorized)
```

```python
predictedValuesFromModel
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,model.predict(X_test_vectorized))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,model.predict(X_test_vectorized))
```

```python
predictedValuesFromModel= np.array(predictedValuesFromModel)
```
