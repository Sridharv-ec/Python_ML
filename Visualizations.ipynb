```python
k=1
plt.figure(figsize=(20,20))
for i in range(1,len(dataset.columns)+1):
    for j in range(1,len(dataset.columns)+1):
        plt.subplot(len(dataset.columns),len(dataset.columns),k)
        k = k+1
        if dataset.columns[i-1] == dataset.columns[j-1]:
            plt.hist(dataset[dataset.columns[i-1]],color=colors[np.random.randint(0,5)])
        else:
            x1=dataset[dataset.columns[i-1]]
            x2=dataset[dataset.columns[j-1]]
            plt.scatter(x1,x2,color=colors[np.random.randint(0,5)])
```
